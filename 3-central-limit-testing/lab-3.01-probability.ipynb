## Exercise 1.2 - Plot probability density

Make a graphical plot of the probability density function and the cumulative distribution function of a normal distribution with $\mu = 2.5$ and $\sigma = 1.5$.

What is the area under the density function between  x=0.5  and  x=4 . (Answer = 0.750)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
# Probability density of a normal distribution with different standard
# deviations
sigma = 1.5
mu = 2.5    # Mean
x = np.linspace(mu-4*sigma, mu+4*sigma, num=1000)

pdf = norm.pdf(x,mu,sigma)
cdf = norm.cdf(x,mu,sigma)

plt.figure(figsize=(10,5))
plt.plot(x,pdf,label="PDF", color="blue")
plt.fill_between(x,pdf,where=(x>0.5)& (x<4),color='skyblue', alpha=0.5)
#plt.plot(x,cdf,label="CDF",color="green")
```

## Exercise 1.3 - Student t vs normal distribution

Make a plot of the probability density function and the cumulative distribution function for the Student's t-distribution with 3 degrees of freedom. 

Also plot the standard normal distribution, so you can compare both.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# Parameters
df = 3  # Degrees of freedom for the Student's t-distribution
x = np.linspace(-5, 5, 1000)  # Range of x values

# Calculate the PDF and CDF for the Student's t-distribution
t_pdf = t.pdf(x, df)
t_cdf = t.cdf(x, df)

# Calculate the PDF and CDF for the standard normal distribution
norm_pdf = norm.pdf(x)
norm_cdf = norm.cdf(x)

# Plot the PDFs
plt.figure(figsize=(12, 6))
plt.plot(x, t_pdf, label="Student's t PDF (df=3)", color='blue')
plt.plot(x, norm_pdf, label='Standard Normal PDF', color='red', linestyle='--')
plt.title("PDFs of Student's t-distribution (df=3) and Standard Normal Distribution")
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# Plot the CDFs
plt.figure(figsize=(12, 6))
plt.plot(x, t_cdf, label="Student's t CDF (df=3)", color='green')
plt.plot(x, norm_cdf, label='Standard Normal CDF', color='orange', linestyle='--')
plt.title("CDFs of Student's t-distribution (df=3) and Standard Normal Distribution")
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()

# Show the plots
plt.show()
```
